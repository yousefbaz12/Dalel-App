```python
import geohash

def add_geohash_feature(data):
    """
     Adds a new column 'Geohash' to the data containing the Geohash code for each place.
    """
    data['Geohash'] = data.apply(lambda row: geohash.encode(row['latitude'], row['longitude']), axis=1)
    return data

gehash_data = add_geohash_feature(data)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import scipy.sparse


# One-hot encode the "Category" column
category_encoder = OneHotEncoder(sparse=False)
category_encoded = category_encoder.fit_transform(gehash_data[['Category']])

# Extract relevant features and perform vectorization (example for text data)
tfidf_vectorizer = TfidfVectorizer(analyzer='word', ngram_range=(1, 2), min_df=0, stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(gehash_data['Reviews List'])

# Add numerical features (latitude, longitude, Rating) and categorical features (Category) to the feature matrix
numerical_features = gehash_data[['latitude', 'longitude', 'Rating']].values
tfidf_matrix_with_numerical = scipy.sparse.hstack((tfidf_matrix, numerical_features), format='csr')

# Combine the TF-IDF matrix and the one-hot encoded categorical features
features_matrix = scipy.sparse.hstack((tfidf_matrix_with_numerical, category_encoded), format='csr')

# Calculate cosine similarity
cosine_sim = cosine_similarity(features_matrix, features_matrix)

# Function to get place index from place name
def get_place_index(place_name, data):
    return data[gehash_data['Place Name'] == place_name].index[0]

def get_recommendations_by_name(place_name, cosine_sim=cosine_sim, data=data):
    """
    Gets recommendations for a given place name, considering both textual similarity and spatial proximity.

    This function assumes the data DataFrame has a 'Geohash' column containing the Geohash code for each place.
    """
    place_index = get_place_index(place_name, data)
    sim_scores = list(enumerate(cosine_sim[place_index]))

    # Sort by a weighted sum of cosine similarity and Geohash similarity
    def weighted_similarity(score):
        place_data = gehash_data.iloc[score[0]]
        geohash_weight = 0.2  # Adjust weight as needed to prioritize location vs. text similarity
        geohash_similarity = sum(int(char == place_data['Geohash'][0]) for char in gehash_data['Geohash'][place_index])
        return score[1] * (1 - geohash_weight) + geohash_similarity * geohash_weight

    sim_scores = sorted(sim_scores, key=weighted_similarity, reverse=True)
    recommended_indices = [i[0] for i in sim_scores[1:4]]  # Get top 3 recommendations
    return gehash_data['Place Name'].iloc[recommended_indices]


# Example: Get recommendations for a place by name
input_place_name = "Tell Basta"
recommendations = get_recommendations_by_name(input_place_name)
print(recommendations)
```
